```python
import bs4 
import requests
import pandas as pd
```

```python
page = 0
name_list = []
data = requests.get('https://www.yellowpages.co.th/heading/%E0%B8%9C%E0%B8%B9%E0%B9%89%E0%B8%AD%E0%B8%AD%E0%B8%81%E0%B9%81%E0%B8%9A%E0%B8%9A%E0%B8%9A%E0%B8%A3%E0%B8%A3%E0%B8%88%E0%B8%B8%E0%B8%A0%E0%B8%B1%E0%B8%93%E0%B8%91%E0%B9%8C?page=' + str(page))
soup = bs4.BeautifulSoup(data.text)
for p in soup.find_all('div',{'class':'col-md-12 col-sm-12 col-xs-12 yp-search-listing no-gutter default-border margin-bottom-fifteen'}):
        name_list.append(p.find('div',{'class':'yp-listing-title col-md-12 col-sm-12 col-xs-12 no-gutter'}).find('a').text)
        
table0 = pd.DataFrame([name_list]).transpose()
table0.columns = ['Company']
table0.set_index('Company')
```
